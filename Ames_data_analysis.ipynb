```python
# importations
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Reading the ames csv file and display the top 5 houses sold
data = pd.read_csv('ames.csv')
data.head()
```

```python
# summary of the whole dataset
data.info()
```

```python
# store the three columns in the variable datasubset
data_subset = data[['SalePrice','TotRmsAbvGrd','OverallCond']]
data_subset.head()
```

```python
# statistical summary of:
data_subset.describe()
```

```python
# Calculate the mean,median and std of saleprice
mean=np.mean(data_subset['SalePrice'])
median=np.median(data_subset['SalePrice'])
Std=data_subset['SalePrice'].std()

# plotting one figure
plt.figure(figsize=(15, 8))
sns.set_style('darkgrid')
plt.subplot(3,2,1)

# visualizing the mean ,median and std of saleprice
# visualizing distribution of sale price
sns.histplot(data=data_subset, x='SalePrice', bins=50,color='brown', kde=True )
plt.title('Sales Distribution')
plt.axvline(mean,color='black',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()

# visualizing saleprice distribution
# Sale price has a lot of outliers
plt.subplot(3,2,2)
sns.boxplot(y='SalePrice',data=data_subset , color='brown')
plt.title('Distribution of SalesPrice')

# getting the mean,meadian and std of TotRmAbvGrd 
mean=np.mean(data_subset['TotRmsAbvGrd'])
median=np.median(data_subset['TotRmsAbvGrd'])
Std=data_subset['TotRmsAbvGrd'].std()

# Visualize TotRmAbvGrd distribution
plt.subplot(3,2,3)
sns.histplot(data=data_subset, x='TotRmsAbvGrd',bins = 14 ,color='green')
plt.title('Distribution of Total Rooms Above Grade')
plt.axvline(mean,color='black',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()

plt.subplot(3,2,4)
sns.boxplot(y='TotRmsAbvGrd',data=data_subset , color='green')
plt.title('Distribution of TotRmsAbvGrd')

# Getting the mean,median and std
mean=np.mean(data_subset['OverallCond'])
median=np.median(data_subset['OverallCond'])
Std=data_subset['OverallCond'].std()

# visualization of overall cond
plt.subplot(3,2,5)
sns.histplot(data=data_subset, x='OverallCond',bins=14 ,color='blue')
plt.title('Distribution of Overall Condition')
plt.axvline(mean,color='black',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()

plt.subplot(3,2,6)
sns.boxplot(y='OverallCond',data=data_subset , color='blue')
plt.title('Distribution of Overall Condition')

plt.tight_layout()
plt.show()S
```

Homes with overall condition of 5 are most sold in Ames
Homes with overall condition of 1,2,3 and 9 are least sold

Homes with total rooms above grade 6 are most sold
Homes with total rooms above grade 3,14,12,11 are least sold

The highest saleprice made from the sale of the homes is $755000
Most homes sold the saleprice is btw 130000-150000$

```python
# Overall condition with total saleprice per condition
Cond_sales = {}
Condition_sales =[['OverallCond','SalePrice']]
for condt in Condition_sales.OverallCond.unique():
    Cond_sales[condt] = [data_subset['SalePrice'][v] for v in data_subset[data_subset["OverallCond"] == condt].index]

Total_sales ={}
for key, value in Cond_sales.items():
    Total_sales[key] = sum(value)

sale_T = Total_sales.items()
Sorted_sales = sorted(sale_T, key=lambda x: x[0])
x = dict(Sorted_sales).keys()
y = dict(Sorted_sales).values()


# #plotting a bar plot for quantity against sales
fig, ax=plt.subplots(figsize=(10,6), ncols=2)

ax[0].bar(x,y)
ax[0].set_ylabel('Total salePrice')
ax[0].set_xlabel('Condition levels')
ax[0].set_title('SalePrice distribution per condition' )

ax[1].scatter(x,y)
ax[1].set_ylabel('Total salePrice')
ax[1].set_xlabel('Condition levels')
ax[1].set_title('SalePrice distribution per condition');

# highest sale price is made from houses under condition 5
# the least sale price is made from houses under condition 1,2,3,9 
```

```python
# subsets based on overall condition
below_average_condition=data[data['OverallCond']<5]
average_condition=data[data['OverallCond']==5]
above_average_condition = data[data['OverallCond']>5]

# use ggplot style
plt.style.use('ggplot')

# plot one figure
plt.figure(figsize=(18,6), layout='constrained')
plt.subplot(2,3,1)

# Create custom bins so all are on the same scale
bins = range(data["SalePrice"].min(), data["SalePrice"].max(), int(data["SalePrice"].median()) // 20) # Plot three histograms, with reduced opacity (alpha) so 

# the three histograms will overlap
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.3,edgecolor='green')
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.5,edgecolor='red')
plt.title('Distribution of saleprice grouped OverallCond')
plt.xlabel('SalePrice')
plt.ylabel('Number of Houses')

# set the legend 
plt.legend(['above_average_condition','average_condition','below_average_condition'])

# set the xticks
xticks = np.arange(50000, 800000, 50000)
plt.xticks(xticks, rotation=90)
plt.show()


# Most sales is made from average condition which is equal to 5
# least sales is made from below_average_condition which is below 5
```

```python
# plot one figure
plt.figure(figsize=(15, 8), layout='constrained')

# subplot three shows how below average condition impacts saleprice
plt.subplot(2,3,1)
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.8,edgecolor='red')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Below average condition')
plt.xticks(rotation=90)

# subplot two shows how average condition impacts saleprice
plt.subplot(2,3,2)
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.8,edgecolor='green')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Average condition')
plt.xticks(rotation=90)

# subplot one shows how above average condition impacts saleprice
plt.subplot(2,3,3)
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.xticks(rotation=90)
plt.title('Above average condition')

plt.show()

# Most saleprice is gotten from houses with average condition
# The amount of saleprice from the houses with below average condition is the least
```

```python
# Correlation
#  storing this 6 columns in a variable data1
data_subset_2=data[['SalePrice','TotRmsAbvGrd','OverallCond','OverallQual','LotArea','LandSlope','YearBuilt','YrSold']]
data_subset_2
```

```python


#Getting the best features that are highly negative and positive correlated to Saleprice 
sns.pairplot(data_subset_2,hue='LandSlope')
plt.show()
# TotRomsAbvGrd and Year Built have a high positive correlation
# with sale price
# YrSold has a high negative correlation with sale price
# from the the pair plot, the prefered houses are those built on gentle slopes hence more sales
```

```python

#Getting the best features that are highly negative and positive correlated to Saleprice  
Corr_data_subset=data_subset_2.corr()

# heatmap to show correlation
sns.heatmap(Corr_data_subset,cmap='coolwarm',vmin=-1,vmax=1,annot=True,square=True,
           annot_kws={'fontsize':11,'fontweight':'bold'})
plt.show()
#  TotRmsAbvGrd positively correlated to sale price
#  OverallCond negatively correlated to sale price.
```

```python
# Saleprice
plt.style.use('fivethirtyeight')
plt.figure(figsize=(16,8))
# plot distribution of columns with the highest positive correlation
plt.subplot(2,2,1)
sns.barplot(y='SalePrice',x='TotRmsAbvGrd',data=data_subset_2)
plt.title('Relationship between saleprice and TotRmsAbvGrd')
plt.plot(legend=False)
# plot distribution of column with the highest positive correlation with saleprice
plt.subplot(2,2,2)
sns.boxplot(y='SalePrice',x='TotRmsAbvGrd', data=data_subset_2)

# Boxplots showing which column is highly correlated with saleprice
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)           
sns.barplot(y='SalePrice',x='OverallCond',data=data_subset_2)
plt.title('Relationship between saleprice and Overall Condition')
plt.subplot(2,2,2)
sns.boxplot(y='SalePrice',x='OverallCond',data=data_subset_2)

plt.show()
```

```python
# creating a new feature age, the difference between Yrsold and YearBuilt
data['Age']=data['YrSold'] - data['YearBuilt']
data.head()
```

```python

 # The relationship btw age and saleprice
plt.figure(figsize=(16,7))
plt.subplot(2,2,1)
scat=sns.scatterplot(y='SalePrice',x='Age',data=data,legend=False,hue='OverallCond',palette='Set2')
scat.set(xlabel='Age of the House')
scat.set_ylabel('SalePrice')
scat.set(title='Relationship between saleprice and Age')
# most saleprice is made from the recent builthouses with age of 0-20years
# The saleprice of Houses of age btw 100-140 is the least

plt.subplot(2,2,2)
sns.boxplot(y='Age',data=data , color='green')
plt.title('Distribution of Age');

# The boxplot shows the median,maximum,minimum and the quartiles of age 
# From the scatter plot, house built recently make more sales than older houses
```
